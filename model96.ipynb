```python
model = Sequential()


model.add(Conv2D(filters=15, 
                 kernel_size=(3,3),
                 padding='same',
                 input_shape=(96,96,1), 
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(30, 30)))
model.add(Flatten()) 

model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=3,activation='softmax'))
adam = adam_v2.Adam(lr=0.0001, beta_1=0.99, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='categorical_crossentropy', 
              optimizer=adam, metrics=['accuracy'])

train_history =model.fit(x=train_feature_vector,
                         y=train_label_onehot,validation_split=0, 
                         epochs=20, batch_size=5,verbose=2)
model.save('a.h5')
model.summary()
"""
scores = model.evaluate(test_feature_vector, test_label_onehot)
print(model.predict(test_feature_vector))
print(test_label_onehot)
print('\n=',scores[1])
model.save('a.h5')
"""
```
